## Q1. Differentiate between Python's `is` and `==` operators.

### Explanation

`==` checks if values are equal. `is` checks if two variables refer to the *exact same object* in memory.

**Demonstration**

```python
list1 = [10, 20, 30]
list2 = [10, 20, 30]
list3 = list1 # list3 points to the same object as list1

print(f"list1 == list2: {list1 == list2}") # True (same content)
print(f"list1 is list2: {list1 is list2}")   # False (different objects)
print(f"list1 is list3: {list1 is list3}")   # True (same object)
```

### Output

```
list1 == list2: True
list1 is list2: False
list1 is list3: True
```

## Q4. Under what circumstances would a tuple be preferred over a list?

### When to use Tuples

Tuples are better when:
1.  **Data should not change:** For fixed collections (e.g., coordinates, database records).
2.  **Used as dictionary keys or set elements:** Tuples are hashable because they are immutable.
3.  **Returning multiple values from a function:** Provides an ordered, unchangeable collection.

## Q5. Explain why dictionary lookups typically achieve O(1) time complexity, whereas searching in a list is O(n).

### Complexity Difference

*   **Dictionaries:** Use hash tables for direct access. On average, finding an item is constant time (O(1)), regardless of dictionary size.
*   **Lists:** Require scanning elements one by one. Finding an item takes linear time (O(n)), proportional to the list's size.

## Q7. Explain the distinctions between `append()`, `extend()`, and `insert()` methods for lists.

Initial list: `my_list = [10, 20, 30]`
Elements to add: `elements_to_add = [40, 50]`

### `append()`
Adds a single element to the end. If you append a list, it adds the list as one item.
```python
current_list = list(my_list)
current_list.append(elements_to_add)
print(f"After append(): {current_list}")
```
Output: `[10, 20, 30, [40, 50]]`

### `extend()`
Adds all elements from an iterable to the end of the list.
```python
current_list = list(my_list)
current_list.extend(elements_to_add)
print(f"After extend(): {current_list}")
```
Output: `[10, 20, 30, 40, 50]`

### `insert()`
Inserts a single element at a specified index.
```python
current_list = list(my_list)
current_list.insert(1, elements_to_add) # Insert at index 1
print(f"After insert(1, ...): {current_list}")
```
Output: `[10, [40, 50], 20, 30]`

````python
# Q27 - Number Diamond Pattern

n = int(input("Enter N : "))

# Upper Half
for i in range(1, n + 1):
    print(" " * (n - i), end="")
    for j in range(1, i + 1):
        print(j, end=" ")
    for j in range(i - 1, 0, -1):
        print(j, end=" ")
    print()

# Lower Half
for i in range(n - 1, 0, -1):
    print(" " * (n - i), end="")
    for j in range(1, i + 1):
        print(j, end=" ")
    for j in range(i - 1, 0, -1):
        print(j, end=" ")
    print()
````

````python
# Q31 - Armstrong Numbers between 1 and 9999

count = 0

print("Armstrong Numbers:")
for num in range(1, 10000):

    temp = num
    digits = len(str(num))
    total = 0
    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10
    if total == num:
        print(num, end=" ")
        count += 1
print("\n")
print("Total Armstrong Numbers =", count)
````

````python
# Q33 - Prime Number and Sum of Prime Numbers

n = int(input("Enter a Number : "))
is_prime = True

if n < 2:
    is_prime = False
else:
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

if is_prime:
    print(n, "is a Prime Number")
else:
    print(n, "is not a Prime Number")

prime_sum = 0

for num in range(2, n + 1):
    prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            prime = False
            break
    if prime:
        prime_sum += num

print("Sum of Prime Numbers =", prime_sum)
````

````python
# Q34 - Reverse Number and Palindrome

number = int(input("Enter Number : "))

if number < 0:
    print("Negative numbers are not palindrome.")
else:
    original = number
    reverse = 0
    while number > 0:
        digit = number % 10
        reverse = reverse * 10 + digit
        number //= 10
    print("Reversed Number =", reverse)

    if original == reverse:
        print("Palindrome")
    else:
        print("Not Palindrome")
````

````python
# Q35 - HCF and LCM

a = int(input("Enter First Number : "))
b = int(input("Enter Second Number : "))

x = a
y = b

while y != 0:
    remainder = x % y
    x = y
    y = remainder

hcf = x
lcm = (a * b) // hcf

print("HCF =", hcf)
print("LCM =", lcm)
````

````python
# Q41 - TCS Stock Price Analysis

prices = (3845, 3902, 3878, 3955, 4012, 3998, 4055)
days = ("Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun")

maximum = max(prices)
minimum = min(prices)

print("Maximum Price :", maximum)
print("Day :", days[prices.index(maximum)])

print()

print("Minimum Price :", minimum)
print("Day :", days[prices.index(minimum)])

average = sum(prices) / len(prices)

print()

print("Average Price :", round(average, 2))

gain = 0
gain_day = ""

for i in range(1, len(prices)):
    difference = prices[i] - prices[i - 1]
    if difference > gain:
        gain = difference
        gain_day = days[i - 1] + " to " + days[i]
print()

print("Biggest Gain :", gain)
print("Between :", gain_day)

volatility = 0

for i in range(1, len(prices)):
    volatility += abs(prices[i] - prices[i - 1])
print("Total Volatility :", volatility)
````

````python
# Q46 - GST Calculator

def calculate_gst(amount, gst_rate=18, gst_inclusive=False):
    if gst_inclusive:
        base = amount / (1 + gst_rate / 100)
        gst = amount - base
        total = amount
    else:
        base = amount
        gst = amount * gst_rate / 100
        total = amount + gst
    return round(base,2), round(gst,2), round(total,2)


print(calculate_gst(1000))
print(calculate_gst(1180, gst_inclusive=True))
print(calculate_gst(500, gst_rate=5))
````

````python
# Q47 - Recursive Power Function

def power(base, exponent):
    if exponent == 0:
        return 1
    elif exponent < 0:
        return 1 / power(base, -exponent)
    else:
        return base * power(base, exponent - 1)

print(power(2,10))
print(power(5,0))
print(power(3,-2))
````
