# Q3.Explain the difference between .fillna() and .dropna(). When would a data analyst pick one over theother? Be specific - give a concrete situation for each.

.fillna() is used to replace missing values with a specified value.
.dropna() is used to remove rows or columns that contain missing values.
A data analyst chooses .fillna() when the missing records are still useful. For example, in a student marks dataset, if a few attendance values are missing, they can be replaced with the average attendance instead of deleting the students records.
A data analyst chooses .dropna() when the missing data is very small and does not affect the analysis. 

For example, if a sales dataset has a few rows with missing product IDs, those rows can simply be removed.

# Q4.What does pd.to_numeric(col, errors='coerce') do that plain pd.to_numeric(col) does NOT? Which one would you use on real-world messy data, and why?

pd.to_numeric(col) converts values to numbers. If it finds an invalid value, it gives an error.
    
pd.to_numeric(col, errors='coerce') also converts values to numbers, but invalid values are changed to NaN instead of giving an error.

I would use errors='coerce' for real-world data because it handles invalid values safely and makes the data easier to clean.

# Q5.Explain .loc vs .iloc with one example each. Then describe a case where df.loc[0] and df.iloc[0] return different rows.

*.loc is used to access data using index labels.

Example:df.loc[0]-Returns the row with index 0.

*.iloc is used to access data using row numbers (positions).

Example:df.iloc[0]-Returns the first row of the DataFrame.

For example, if the index is [2, 0, 1]:

df.loc[0] returns the row with index 0.

df.iloc[0] returns the first row (index 2).

so,.loc uses labels, while .iloc uses positions.

# Q6.What is the difference between .map(), .apply(), and .replace() for a Series? When would you reach for each?

.map() is used to change values in a Series.

.apply() is used to perform a function on each value.

.replace() is used to replace one value with another.

Example:
.map() → Change "M" to "Male".
.apply() → Multiply every number by 2.
.replace() → Replace "N/A" with "Unknown".
